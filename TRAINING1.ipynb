```python
%load_ext autoreload
%autoreload 2
import qiskit_metal as metal
from qiskit_metal import designs, draw
from qiskit_metal import MetalGUI, Dict
```

```python
design = designs.DesignPlanar()
design.overwrite_enabled = True
gui = MetalGUI(design)
```

#### Creating Qubits

```python
from qiskit_metal.qlibrary.qubits.transmon_pocket import TransmonPocket

options = Dict(
    connection_pads = Dict(
        a = dict(loc_W=+1,loc_H=+1),
        b = dict(loc_W=-1,loc_H=+1),
        c = dict(loc_W=+1,loc_H=-1),
        d = dict(loc_W=-1,loc_H=-1)
    ),
    pos_x = "-2.5mm",
    pos_y = '0mm',
    orientation = '0'
)

# Qubit 1
q_1 = TransmonPocket(design=design, name="Q1", options=options)

# Qubit 2 
q_2 = TransmonPocket(design=design, name="Q2", options=options)
q_2.options.pos_x = '2.5mm'

# Qubit 3
q_3 = TransmonPocket(design=design, name="Q3", options=options)
q_3.options.pos_x = '0mm'
q_3.options.pos_y = '1mm'
q_3.options.orientation = '90'

# Qubit 4
q_4 = TransmonPocket(design=design, name="Q4", options=options)
q_4.options.pos_x = '0mm'
q_4.options.pos_y = '-1mm'
q_4.options.orientation = '90'


gui.rebuild()
gui.autoscale()
```

#### Adding CPW

```python
from qiskit_metal.qlibrary.tlines.meandered import RouteMeander
RouteMeander.get_template_options(design)
```

```python
# connecting qubits 1 and 3
# we want to connect pins a in qubit 1 with pin b in qubit 3
option13 = Dict(
    pin_inputs = Dict(
        start_pin=Dict(
            component="Q1",
            pin="a"
        ),
        end_pin= Dict(
            component="Q3",
            pin="b"
        )
    ),
    trace_width='10um',
    trace_gap='6um',
    lead=Dict(
        start_straight='125um',
        end_straight='125um'
    ),
    fillet='90um',
    meander= Dict(asymmetry='500um'),
    total_length='6mm'
)

cpw_1_3 = RouteMeander(design=design, name="CPW_1_3", options=option13)

gui.rebuild()
gui.autoscale()
```

```python
# connecting qubits 3 and 2
# we want to connect pins d in qubit 3 with pin b in qubit 2
option23 = Dict(
    pin_inputs = Dict(
        start_pin=Dict(
            component="Q2",
            pin="b"
        ),
        end_pin= Dict(
            component="Q3",
            pin="d"
        )
    ),
    trace_width='10um',
    trace_gap='6um',
    lead=Dict(
        start_straight='125um',
        end_straight='125um'
    ),
    fillet='90um',
    meander= Dict(asymmetry='-500um'),
    total_length='6mm'
)

cpw_2_3 = RouteMeander(design=design, name="CPW_2_3", options=option23)

gui.rebuild()
gui.autoscale()
```

```python
# connecting qubits 1 and 4
# we want to connect pins c in qubit 3 with pin a in qubit 4
option14 = Dict(
    pin_inputs = Dict(
        start_pin=Dict(
            component="Q1",
            pin="c"
        ),
        end_pin= Dict(
            component="Q4",
            pin="a"
        )
    ),
    trace_width='10um',
    trace_gap='6um',
    lead=Dict(
        start_straight='125um',
        end_straight='125um'
    ),
    fillet='90um',
    meander= Dict(asymmetry='-500um'),
    total_length='6mm'
)

cpw_1_4 = RouteMeander(design=design, name="CPW_1_4", options=option14)

gui.rebuild()
gui.autoscale()
```

```python
# connecting qubits 2 and 4
# we want to connect pins c in qubit 3 with pin c in qubit 4
option24 = Dict(
    pin_inputs = Dict(
        start_pin=Dict(
            component="Q2",
            pin="d"
        ),
        end_pin= Dict(
            component="Q4",
            pin="c"
        )
    ),
    trace_width='10um',
    trace_gap='6um',
    lead=Dict(
        start_straight='125um',
        end_straight='125um'
    ),
    fillet='90um',
    meander= Dict(asymmetry='500um'),
    total_length='6mm'
)

cpw_2_4 = RouteMeander(design=design, name="CPW_2_4", options=option24)

gui.rebuild()
gui.autoscale()
```

#### Adding Launchers

According to the design provided on the website we will add 8 launchers

```python
from qiskit_metal.qlibrary.terminations.launchpad_wb import LaunchpadWirebond
from qiskit_metal.qlibrary.terminations.launchpad_wb_coupled import LaunchpadWirebondCoupled

# Corners
l_1_c = LaunchpadWirebond(design=design, name="L_1_C", options=Dict(pos_x='-3545um', pos_y='-2812um', orientation='90', lead_length='0um'))
l_2_c = LaunchpadWirebond(design=design, name="L_2_C", options=Dict(pos_x='3545um', pos_y='2812um', orientation='270', lead_length='0um'))
l_3_c = LaunchpadWirebond(design=design, name="L_3_C", options=Dict(pos_x='-3545um', pos_y='2812um', orientation='270', lead_length='0um'))
l_4_c = LaunchpadWirebond(design=design, name="L_4_C", options=Dict(pos_x='3545um', pos_y='-2812um', orientation='90', lead_length='0um'))

gui.rebuild()
gui.autoscale()
```

```python

# V2 of launchers
l_1_q = LaunchpadWirebond(design=design, name="L_1_Q", options=Dict(pos_x='-4020um', pos_y='0um', orientation='0', lead_length='30um'))
l_2_q = LaunchpadWirebond(design=design, name="L_2_Q", options=Dict(pos_x='4020um', pos_y='0um', orientation='180', lead_length='30um'))
l_3_q = LaunchpadWirebond(design=design, name="L_3_Q", options=Dict(pos_x='990um', pos_y='2812um', orientation='270', lead_length='30um'))
l_4_q = LaunchpadWirebond(design=design, name="L_4_Q", options=Dict(pos_x='-990um', pos_y='-2812um', orientation='90', lead_length='30um'))

gui.rebuild()
gui.autoscale()
```

```python
# The Q launchers will be connected to the qubits with meander structures
# for q1
option_lq1 = Dict(
    pin_inputs = Dict(
        start_pin=Dict(
            component="Q1",
            pin="b"
        ),
        end_pin= Dict(
            component="L_1_Q",
            pin="tie"
        )
    ),
    trace_width='10um',
    trace_gap='6um',
    lead=Dict(
        start_straight='430um',
        end_straight='0um'
    ),
    fillet='90um',
    meander= Dict(asymmetry='150um'),
    total_length='6mm'
)

lq1 = RouteMeander(design=design, name="LQ1", options=option_lq1)

# for q2
option_lq2 = Dict(
    pin_inputs = Dict(
        start_pin=Dict(
            component="L_2_Q",
            pin="tie"
        ),
        end_pin= Dict(
            component="Q2",
            pin="c"
        )
    ),
    trace_width='10um',
    trace_gap='6um',
    lead=Dict(
        start_straight='430um',
        end_straight='0um'
    ),
    fillet='90um',
    meander= Dict(asymmetry='150um'),
    total_length='6mm'
)

lq2 = RouteMeander(design=design, name="LQ2", options=option_lq2)

# for q3
option_lq3 = Dict(
    pin_inputs = Dict(
        start_pin=Dict(
            component="Q3",
            pin="c"
        ),
        end_pin= Dict(
            component="L_3_Q",
            pin="tie"
        )
    ),
    trace_width='10um',
    trace_gap='6um',
    lead=Dict(
        start_straight='430um',
        end_straight='0um'
    ),
    fillet='90um',
    meander= Dict(asymmetry='50um'),
    total_length='6mm'
)

lq3 = RouteMeander(design=design, name="LQ3", options=option_lq3)


# forq4
# for q3
option_lq4 = Dict(
    pin_inputs = Dict(
        start_pin=Dict(
            component="L_4_Q",
            pin="tie"
        ),
        end_pin= Dict(
            component="Q4",
            pin="b"
        )
    ),
    trace_width='10um',
    trace_gap='6um',
    lead=Dict(
        start_straight='430um',
        end_straight='0um'
    ),
    fillet='90um',
    meander= Dict(asymmetry='50um'),
    total_length='6mm'
)

lq4 = RouteMeander(design=design, name="LQ4", options=option_lq4)

gui.rebuild()
gui.autoscale()
```

```python
# corner launchers will be connected to the qubit without any  meander structure
from qiskit_metal.qlibrary.tlines.anchored_path import RouteAnchors
from collections import OrderedDict
jogsA_in = OrderedDict()
jogsA_in[0] = ["L", '200um']

options_line_cl1 = {'pin_inputs':
            {'start_pin': {'component': 'Q1', 'pin': 'd'},
             'end_pin': {'component': 'L_1_C', 'pin': 'tie'}},
            'lead': {'start_straight': '120um', 'end_straight': '225um', 'start_jogged_extension': jogsA_in},
            'fillet': '100um'
            }
cl1 = RouteAnchors(design=design, name='line_cl1', options= options_line_cl1)

options_line_cl3 = {'pin_inputs':
            {'start_pin': {'component': 'Q3', 'pin': 'a'},
             'end_pin': {'component': 'L_3_C', 'pin': 'tie'}},
            'lead': {'start_straight': '120um', 'end_straight': '225um', 'start_jogged_extension': jogsA_in},
            'fillet': '100um'
            }
cl3 = RouteAnchors(design, 'line_cl3', options_line_cl3)

gui.rebuild()
gui.autoscale()
```

```python
jogsB_in = OrderedDict()
jogsB_in[0] = ["L", '300um']

options_line_cl2 = {'pin_inputs':
            {'start_pin': {'component': 'Q2', 'pin': 'a'},
             'end_pin': {'component': 'L_2_C', 'pin': 'tie'}},
            'lead': {'start_straight': '120um', 'end_straight': '225um', 'start_jogged_extension': jogsB_in},
            'fillet': '100um'
            }
cl2 = RouteAnchors(design=design, name='line_cl2', options= options_line_cl2)

options_line_cl4 = {'pin_inputs':
            {'start_pin': {'component': 'Q4', 'pin': 'd'},
             'end_pin': {'component': 'L_4_C', 'pin': 'tie'}},
            'lead': {'start_straight': '120um', 'end_straight': '225um', 'start_jogged_extension': jogsB_in},
            'fillet': '100um'
            }
cl4 = RouteAnchors(design, 'line_cl4', options_line_cl4)

gui.rebuild()
gui.autoscale()
```
